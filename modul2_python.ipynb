### 2.1 Variabel & Assignment

Variabel adalah tempat menyimpan data. Bayangkan seperti kotak yang isinya bisa berubah-ubah.

**Aturan penamaan variabel:**
- Gunakan huruf, angka, dan underscore (_)
- Dimulai dengan huruf atau underscore, bukan angka
- Hindari spasi (gunakan _ untuk pemisah)
- Hindari kata kunci Python (seperti if, for, class)

Contoh yang BENAR:
- nama = "Budi"
- angkatan_2023 = 2023
- ipk = 3.75

Contoh yang SALAH:
- 123nama = "Budi" (dimulai dengan angka)
- nama lengkap = "Budi Santoso" (ada spasi)


````python
# Contoh Variabel dan Assignment
nama = "Budi Wijaya"
nim = "1301213001"
angkatan = 2023
ipk = 3.75
aktif = True

# Cetak nilai variabel
print(f"Nama: {nama}")
print(f"NIM: {nim}")
print(f"Angkatan: {angkatan}")
print(f"IPK: {ipk}")
print(f"Aktif: {aktif}")
````

### 2.2 Tipe Data di Python

Python memiliki beberapa tipe data dasar:

| Tipe | Contoh | Deskripsi |
|------|--------|----------|
| **String** | "Budi", '2024' | Teks |
| **Integer** | 2023, -5 | Bilangan bulat |
| **Float** | 3.75, 2.5 | Bilangan desimal |
| **Boolean** | True, False | Nilai benar/salah |

Kita bisa mengecek tipe data dengan fungsi type() dan mengkonversi dengan int(), str(), float().


````python
# Contoh berbagai tipe data
nama_mk = "Basis Data"  # String
sks = 3                  # Integer
durasi_jam = 2.5       # Float
tersedia = True        # Boolean

# Cek tipe data
print("Tipe data:")
print(f"nama_mk: {type(nama_mk)}")
print(f"sks: {type(sks)}")
print(f"durasi_jam: {type(durasi_jam)}")
print(f"tersedia: {type(tersedia)}")

# Konversi tipe data
sks_string = str(sks)
print(f"\nSKS sebagai string: {sks_string}, tipe: {type(sks_string)}")

durasi_int = int(durasi_jam)
print(f"Durasi sebagai integer: {durasi_int}, tipe: {type(durasi_int)}")
````

### 2.3 Operasi Dasar

#### Operasi Aritmetika
- `+` : Tambah
- `-` : Kurang
- `*` : Kali
- `/` : Bagi (hasil desimal)
- `//` : Bagi bulat (floor division)
- `%` : Modulus (sisa pembagian)
- `**` : Pangkat

#### Operasi String
- `+` : Concatenation (gabung)
- `*` : Repetition (ulang)
- `.upper()` : Ubah ke UPPERCASE
- `.lower()` : Ubah ke lowercase
- `len()` : Panjang string

#### Operasi Perbandingan
- `==` : Sama dengan
- `!=` : Tidak sama dengan
- `>` : Lebih besar
- `<` : Lebih kecil
- `>=` : Lebih besar atau sama
- `<=` : Lebih kecil atau sama


````python
# Operasi Aritmetika
total_sks = 3 + 4 + 2  # Tambah SKS
print(f"Total SKS: {total_sks}")

sisa_jam = 20 % 3  # Berapa sisa jika 20 dibagi 3
print(f"20 % 3 = {sisa_jam}")

durasi = 2 ** 3  # 2 pangkat 3
print(f"2^3 = {durasi}")

# Operasi String
greeting = "Halo " + "Budi"
print(f"\n{greeting}")

bintang = "*" * 10
print(bintang)

nama_besar = "budi".upper()
print(f"Nama uppercase: {nama_besar}")

# Operasi Perbandingan
print(f"\n2023 == 2024: {2023 == 2024}")
print(f"IPK 3.75 >= 3.0: {3.75 >= 3.0}")
````

### 2.4 Print & f-string

Fungsi print() digunakan untuk menampilkan output.


Kita juga bisa format angka:

### 2.5 List - Kumpulan Data

List adalah kumpulan data yang disimpan dalam urutan tertentu. List menggunakan tanda [].

```python
daftar_mhs = ["Budi", "Ani", "Citra"]
nilai_ujian = [85, 90, 78, 92]
```

**Operasi List:**
- **Akses**: `daftar_mhs[0]` → "Budi" (index mulai dari 0)
- **Akses negatif**: `daftar_mhs[-1]` → "Citra" (elemen terakhir)
- **Slicing**: `daftar_mhs[0:2]` → ["Budi", "Ani"] (dari index 0 sampai sebelum 2)
- **Method**: `.append()`, `.remove()`, `.sort()`, `len()`
- **Iterasi**: Loop melalui setiap elemen


````python
# Membuat list
daftar_mhs = ["Budi Wijaya", "Ani Kusuma", "Citra Pratama", "Doni Rahman"]
nilai_ujian = [85, 90, 78, 92]

# Akses elemen
print(f"Mahasiswa pertama: {daftar_mhs[0]}")
print(f"Mahasiswa terakhir: {daftar_mhs[-1]}")

# Panjang list
print(f"\nJumlah mahasiswa: {len(daftar_mhs)}")

# Slicing
print(f"Dua mahasiswa pertama: {daftar_mhs[0:2]}")

# Tambah elemen
daftar_mhs.append("Elsa Hermawan")
print(f"\nSetelah tambah: {daftar_mhs}")

# Iterasi list
print("\nNilai ujian:")
for nilai in nilai_ujian:
    print(f"  - {nilai}")
````

### 2.6 Dictionary - Data Terstruktur

Dictionary menyimpan data dengan pasangan **key: value**. Dictionary menggunakan tanda {}.

```python
mhs = {
    "nim": "1301213001",
    "nama": "Budi Wijaya",
    "kelas": "IF-A",
    "ipk": 3.75
}
```

**Operasi Dictionary:**
- **Akses**: `mhs["nama"]` → "Budi Wijaya"
- **Get dengan default**: `mhs.get("angkatan", "Tidak ada")` → "Tidak ada" (jika key tidak ada)
- **Update**: `mhs["ipk"] = 3.80`
- **Method**: `.keys()`, `.values()`, `.items()`

**List of Dictionaries** - Representasi tabel/banyak baris data:
```python
data_mhs = [
    {"nim": "1301213001", "nama": "Budi", "ipk": 3.75},
    {"nim": "1301213002", "nama": "Ani", "ipk": 3.80},
]
```


````python
# Membuat dictionary
mhs1 = {
    "nim": "1301213001",
    "nama": "Budi Wijaya",
    "kelas": "IF-A",
    "ipk": 3.75
}

# Akses nilai
print(f"Nama: {mhs1['nama']}")
print(f"NIM: {mhs1['nim']}")

# Get dengan default
angkatan = mhs1.get("angkatan", "2023")
print(f"Angkatan: {angkatan}")

# Update nilai
mhs1["ipk"] = 3.80
print(f"IPK (updated): {mhs1['ipk']}")

# List of Dictionaries
data_mhs = [
    {"nim": "1301213001", "nama": "Budi", "kelas": "IF-A"},
    {"nim": "1301213002", "nama": "Ani", "kelas": "IF-B"},
    {"nim": "1301213003", "nama": "Citra", "kelas": "SI-A"}
]

print(f"\nData {len(data_mhs)} mahasiswa:")
for mhs in data_mhs:
    print(f"  {mhs['nama']} ({mhs['nim']}) - {mhs['kelas']}")
````

## Bagian 3: Control Flow

### 3.1 If-Elif-Else

Digunakan untuk membuat keputusan dalam program.

```python
if kondisi:
    # Jalankan jika kondisi TRUE
elif kondisi_lain:
    # Jalankan jika kondisi_lain TRUE
else:
    # Jalankan jika semua FALSE
```


````python
# Contoh 1: Cek angkatan
angkatan = 2022
tahun_sekarang = 2024

tahun_kuliah = tahun_sekarang - angkatan

if tahun_kuliah <= 2:
    kategori = "Junior"
elif tahun_kuliah <= 3:
    kategori = "Menengah"
else:
    kategori = "Senior"

print(f"Angkatan {angkatan} → {kategori}")

# Contoh 2: Labelisasi beban SKS
total_sks = 15

if total_sks < 12:
    beban = "Ringan"
elif total_sks <= 20:
    beban = "Normal"
else:
    beban = "Berat"

print(f"Total SKS {total_sks} → Beban: {beban}")
````

### 3.2 For Loop

Loop digunakan untuk mengulangi suatu blok kode.

```python
for item in list:
    # Jalankan untuk setiap item
```


````python
# Iterasi list
daftar_mhs = ["Budi", "Ani", "Citra"]
for nama in daftar_mhs:
    print(f"Nama: {nama}")

# Iterasi list of dictionaries
print("\nData Mata Kuliah:")
data_mk = [
    {"nama": "Basis Data", "sks": 3},
    {"nama": "Algoritma", "sks": 4},
    {"nama": "Web", "sks": 2}
]

for mk in data_mk:
    print(f"{mk['nama']} - {mk['sks']} SKS")

# Menggunakan range()
print("\nBilangan 1-5:")
for i in range(1, 6):
    print(i)

# Menggunakan enumerate()
print("\nDaftar dengan index:")
for index, nama in enumerate(daftar_mhs):
    print(f"{index+1}. {nama}")
````

### 3.3 Fungsi (def)

Fungsi adalah blok kode yang dapat digunakan kembali.

```python
def nama_fungsi(parameter1, parameter2):
    # Lakukan sesuatu
    return hasil
```


````python
# Fungsi hitung total SKS
def hitung_total_sks(list_sks):
    """Hitung total SKS dari list angka"""
    total = 0
    for sks in list_sks:
        total += sks
    return total

# Fungsi labelisasi beban
def label_beban(total_sks):
    """Labeli beban SKS sebagai Ringan/Normal/Berat"""
    if total_sks < 12:
        return "Ringan"
    elif total_sks <= 20:
        return "Normal"
    else:
        return "Berat"

# Test fungsi
sks_list = [3, 4, 2, 3]
total = hitung_total_sks(sks_list)
beban = label_beban(total)

print(f"Total SKS: {total}")
print(f"Beban: {beban}")
````

## Bagian 4: Error Handling

### 4.1 Apa itu Error?

Error terjadi ketika program menemukan kondisi yang tidak diharapkan. Beberapa error umum:

- **NameError** - Variabel tidak ditemukan
- **TypeError** - Tipe data tidak sesuai
- **KeyError** - Key dictionary tidak ada
- **ValueError** - Konversi tipe gagal
- **ZeroDivisionError** - Pembagian dengan nol


````python
# DEMO ERROR - Jalankan satu per satu untuk lihat error

# Contoh TypeError - bisa langsung dilihat
try:
    hasil = "3" + 5  # String + Integer
except TypeError as e:
    print(f"TypeError: {e}")

# Contoh ValueError
try:
    angka = int("bukan_angka")
except ValueError as e:
    print(f"ValueError: {e}")
````

### 4.2 Try-Except

Try-except digunakan untuk menangani error tanpa program crash.

```python
try:
    # Kode yang mungkin error
except TipeError:
    # Jalankan jika error terjadi
```


````python
# Contoh 1: Konversi String ke Integer
def konversi_ke_int(value):
    try:
        return int(value)
    except ValueError:
        print(f"Error: '{value}' bukan angka!")
        return None

hasil1 = konversi_ke_int("123")
hasil2 = konversi_ke_int("tiga")

print(f"Hasil 1: {hasil1}")
print(f"Hasil 2: {hasil2}")
````

### 4.3 Try-Except-Else-Finally

Pattern lengkap untuk error handling:

```python
try:
    # Kode yang mungkin error
except ErrorType:
    # Jalankan jika error terjadi
else:
    # Jalankan jika TRY BERHASIL (tidak ada error)
finally:
    # Jalankan selalu (ada/tidak ada error)
```


````python
# Contoh try-except-else-finally
def baca_dan_proses(data_str):
    try:
        nilai = float(data_str)
    except ValueError:
        print(f"Gagal: '{data_str}' bukan angka")
    else:
        # Jalankan jika konversi berhasil
        print(f"Sukses: Nilai {nilai}")
        if nilai < 0:
            print("  - Nilai negatif")
        elif nilai >= 80:
            print("  - Nilai Bagus!")
    finally:
        # Jalankan selalu
        print("  - Proses selesai\n")

baca_dan_proses("85.5")
baca_dan_proses("tiga")
````
