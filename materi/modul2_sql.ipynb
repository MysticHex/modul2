## Bagian 6: SQLite Database

### 6.1 Pengenalan SQLite

SQLite adalah database relasional ringan yang built-in di Python. Cocok untuk pembelajaran karena:
- Tidak perlu install server terpisah
- Semua data tersimpan dalam 1 file .db
- Mendukung SQL queries


```python
# Import sqlite3
import sqlite3

# Koneksi ke database (akan membuat file akademik.db jika belum ada)
conn = sqlite3.connect("akademik.db")
print("Database akademik.db berhasil dibuat/dibuka")
```

### 6.2 Load Data CSV ke SQLite

Gunakan Pandas untuk menulis DataFrame ke tabel SQL dengan method `.to_sql()`


```python
# Load data ke database
df_mahasiswa.to_sql("mahasiswa", conn, if_exists="replace", index=False)
df_mata_kuliah.to_sql("mata_kuliah", conn, if_exists="replace", index=False)
df_jadwal.to_sql("jadwal", conn, if_exists="replace", index=False)

print("Data berhasil di-load ke database!")
```

### 6.3 Verifikasi Data di Database


```python
# Query sederhana untuk verifikasi
query = "SELECT * FROM mahasiswa LIMIT 5"
result = pd.read_sql_query(query, conn)
print("5 Mahasiswa pertama dari database:")
print(result)
```

## Bagian 7: SQL for Data Analysis

### 7.1 SELECT & WHERE

Query dasar SQL untuk mengambil dan memfilter data.


```python
# Query 1: Ambil semua mahasiswa
query = "SELECT * FROM mahasiswa"
result = pd.read_sql_query(query, conn)
print(f"Jumlah mahasiswa: {len(result)}")

# Query 2: Hanya kolom tertentu
query = "SELECT nama_lengkap, kelas, angkatan FROM mahasiswa"
result = pd.read_sql_query(query, conn)
print("\nNama, Kelas, Angkatan:")
print(result.head())
```

```python
# Query 3: Filter dengan WHERE
query = "SELECT nama_lengkap, kelas FROM mahasiswa WHERE angkatan = 2023"
result = pd.read_sql_query(query, conn)
print(f"Mahasiswa Angkatan 2023: {len(result)} orang")
print(result)
```

```python
# Query 4: Mata kuliah dengan SKS >= 3
query = "SELECT * FROM mata_kuliah WHERE sks >= 3"
result = pd.read_sql_query(query, conn)
print(f"Mata Kuliah dengan SKS >= 3: {len(result)}")
print(result)
```

### 7.2 ORDER BY & DISTINCT


```python
# Query: Mahasiswa diurutkan berdasarkan nama
query = "SELECT nama_lengkap, kelas FROM mahasiswa ORDER BY nama_lengkap"
result = pd.read_sql_query(query, conn)
print("Mahasiswa (urut alfabet):")
print(result.head(10))
```

```python
# Query: Daftar kelas unik
query = "SELECT DISTINCT kelas FROM mahasiswa ORDER BY kelas"
result = pd.read_sql_query(query, conn)
print("Daftar Kelas Unik:")
print(result)
```

### 7.3 Aggregation - COUNT, SUM, AVG


```python
# Query 1: Jumlah mahasiswa per kelas
query = "SELECT kelas, COUNT(*) as jumlah FROM mahasiswa GROUP BY kelas"
result = pd.read_sql_query(query, conn)
print("Jumlah Mahasiswa per Kelas:")
print(result)
```

```python
# Query 2: Rata-rata SKS per mata kuliah
query = "SELECT AVG(sks) as rata_rata_sks FROM mata_kuliah"
result = pd.read_sql_query(query, conn)
print("Rata-rata SKS Mata Kuliah:")
print(result)
```

```python
# Query 3: Total SKS per angkatan
query = """
SELECT angkatan, COUNT(*) as jumlah_mhs
FROM mahasiswa
WHERE angkatan IS NOT NULL
GROUP BY angkatan
"""
result = pd.read_sql_query(query, conn)
print("Jumlah Mahasiswa per Angkatan:")
print(result)
```

### 7.4 JOIN - Menggabungkan Tabel

JOIN digunakan untuk menggabungkan data dari tabel yang berbeda.


```python
# Query: INNER JOIN (3 tabel)
query = """
SELECT 
    m.nama_lengkap,
    mk.nama_mk,
    mk.sks,
    j.hari,
    j.jam_mulai,
    j.ruangan
FROM jadwal j
INNER JOIN mahasiswa m ON j.mahasiswa_id = m.mahasiswa_id
INNER JOIN mata_kuliah mk ON j.mk_id = mk.mk_id
LIMIT 10
"""
result = pd.read_sql_query(query, conn)
print("Jadwal Kuliah (INNER JOIN):")
print(result)
```

```python
# Query: LEFT JOIN (untuk lihat mahasiswa tanpa jadwal)
query = """
SELECT 
    m.nama_lengkap,
    COUNT(j.jadwal_id) as jumlah_mk
FROM mahasiswa m
LEFT JOIN jadwal j ON m.mahasiswa_id = j.mahasiswa_id
GROUP BY m.mahasiswa_id, m.nama_lengkap
ORDER BY jumlah_mk
"""
result = pd.read_sql_query(query, conn)
print("Jumlah Mata Kuliah per Mahasiswa (termasuk yang belum ada):")
print(result)
```

### 7.5 Studi Kasus KPI (Key Performance Indicator)


```python
# KPI 1: Mahasiswa yang belum punya jadwal (Left join)
query = """
SELECT m.nama_lengkap, m.kelas, m.angkatan
FROM mahasiswa m
LEFT JOIN jadwal j ON m.mahasiswa_id = j.mahasiswa_id
WHERE j.jadwal_id IS NULL
"""
result = pd.read_sql_query(query, conn)
print("Mahasiswa tanpa Jadwal:")
print(result)
```

```python
# KPI 2: Mata kuliah yang belum diambil siapa pun
query = """
SELECT mk.nama_mk, mk.sks, mk.dosen_pengampu
FROM mata_kuliah mk
LEFT JOIN jadwal j ON mk.mk_id = j.mk_id
WHERE j.jadwal_id IS NULL
"""
result = pd.read_sql_query(query, conn)
print("Mata Kuliah yang belum diambil:")
print(result)
```

## Bagian 8: Integrasi SQL + Python

### 8.1 Tampung Hasil Query ke Variabel Python


```python
# Query data
query = "SELECT nama_lengkap, kelas, angkatan FROM mahasiswa LIMIT 5"
result_df = pd.read_sql_query(query, conn)

# Konversi ke list
list_nama = result_df["nama_lengkap"].tolist()
print(f"Sebagai List: {list_nama}")

# Konversi ke list of dictionaries
list_dict = result_df.to_dict("records")
print(f"\nSebagai List of Dict: {list_dict[0]}")
```

### 8.2 Labelisasi dengan Control Flow


```python
# Query total SKS per mahasiswa
query = """
SELECT 
    m.nama_lengkap,
    COUNT(j.mk_id) as jumlah_mk,
    COALESCE(SUM(mk.sks), 0) as total_sks
FROM mahasiswa m
LEFT JOIN jadwal j ON m.mahasiswa_id = j.mahasiswa_id
LEFT JOIN mata_kuliah mk ON j.mk_id = mk.mk_id
GROUP BY m.mahasiswa_id
LIMIT 10
"""

result = pd.read_sql_query(query, conn)

# Labelisasi dengan Python
def label_beban_sks(total_sks):
    if pd.isna(total_sks) or total_sks < 12:
        return "Ringan"
    elif total_sks <= 20:
        return "Normal"
    else:
        return "Berat"

result["beban"] = result["total_sks"].apply(label_beban_sks)

print("Beban SKS Mahasiswa:")
print(result)
```

### 8.3 Validasi Data


```python
# Validasi 1: Cek data null di tabel mahasiswa
print("=== DATA NULL DI MAHASISWA ===")
for col in df_mahasiswa.columns:
    null_count = df_mahasiswa[col].isnull().sum()
    if null_count > 0:
        print(f"{col}: {null_count} null")

# Validasi 2: Cek referential integrity (mk_id di jadwal ada di mata_kuliah?)
print("\n=== REFERENTIAL INTEGRITY CHECK ===")
query = """
SELECT COUNT(*) as invalid_mk
FROM jadwal
WHERE mk_id NOT IN (SELECT mk_id FROM mata_kuliah)
"""
result = pd.read_sql_query(query, conn)
invalid_count = result["invalid_mk"][0]
print(f"Jadwal dengan mk_id invalid: {invalid_count}")
```

### 8.4 Fungsi Lengkap: Query + Proses + Return


```python
def laporan_mahasiswa(mahasiswa_id, conn):
    """
    Generate laporan lengkap untuk satu mahasiswa
    """
    # 1. Query data mahasiswa
    query_mhs = f"""
    SELECT * FROM mahasiswa WHERE mahasiswa_id = {mahasiswa_id}
    """
    result_mhs = pd.read_sql_query(query_mhs, conn)
    
    if len(result_mhs) == 0:
        return {"error": "Mahasiswa tidak ditemukan"}
    
    mhs = result_mhs.iloc[0]
    
    # 2. Query jadwal lengkap
    query_jadwal = f"""
    SELECT mk.nama_mk, mk.sks, j.hari, j.jam_mulai
    FROM jadwal j
    JOIN mata_kuliah mk ON j.mk_id = mk.mk_id
    WHERE j.mahasiswa_id = {mahasiswa_id}
    """
    result_jadwal = pd.read_sql_query(query_jadwal, conn)
    
    # 3. Hitung total SKS (handle string values)
    total_sks = 0
    try:
        for sks in result_jadwal["sks"]:
            try:
                total_sks += float(sks)
            except (ValueError, TypeError):
                # Skip non-numeric SKS values
                pass
    except:
        total_sks = 0
    
    # 4. Labelisasi
    if total_sks < 12:
        beban = "Ringan"
    elif total_sks <= 20:
        beban = "Normal"
    else:
        beban = "Berat"
    
    # 5. Return laporan
    return {
        "nama": mhs["nama_lengkap"],
        "nim": mhs["nim"],
        "kelas": mhs["kelas"],
        "total_sks": int(total_sks),
        "beban": beban,
        "jumlah_mk": len(result_jadwal)
    }

# Test fungsi
laporan = laporan_mahasiswa(1, conn)
print(f"Laporan Mahasiswa:")
for key, value in laporan.items():
    print(f"  {key}: {value}")
```

## Bagian 9: Mini Project

### Judul: "Laporan Akademik Otomatis"

**Instruksi:**
1. Query seluruh data jadwal lengkap (JOIN 3 tabel)
2. Tampung ke variabel Python
3. Untuk setiap mahasiswa:
   - Hitung total SKS
   - Beri label beban SKS
   - Cek apakah ada data tidak valid
4. Cetak laporan ringkasan dalam format yang rapi
5. Handle semua kemungkinan error


```python
# Mini Project: Laporan Akademik Otomatis

# Step 1: Query seluruh jadwal (dengan validasi)
query = """
SELECT 
    m.mahasiswa_id,
    m.nama_lengkap,
    m.kelas,
    mk.nama_mk,
    mk.sks,
    j.hari,
    j.jam_mulai,
    CASE WHEN j.ruangan IS NULL THEN 'INVALID' ELSE j.ruangan END as ruangan
FROM jadwal j
INNER JOIN mahasiswa m ON j.mahasiswa_id = m.mahasiswa_id
LEFT JOIN mata_kuliah mk ON j.mk_id = mk.mk_id
ORDER BY m.mahasiswa_id
"""

data_jadwal = pd.read_sql_query(query, conn)

print(f"Total jadwal valid: {len(data_jadwal)}")
print("\nContoh data:")
print(data_jadwal.head(10))
```

```python
# Step 2-5: Process dan generate laporan

laporan_list = []

# Group by mahasiswa
for mhs_id in data_jadwal["mahasiswa_id"].unique():
    try:
        mhs_data = data_jadwal[data_jadwal["mahasiswa_id"] == mhs_id]
        
        # Ambil info dasar
        nama = mhs_data["nama_lengkap"].iloc[0]
        kelas = mhs_data["kelas"].iloc[0]
        
        # Hitung total SKS
        try:
            total_sks = mhs_data[mhs_data["sks"].notna()]["sks"].astype(float).sum()
        except:
            total_sks = 0
        
        # Label beban
        if total_sks < 12:
            beban = "Ringan"
        elif total_sks <= 20:
            beban = "Normal"
        else:
            beban = "Berat"
        
        # Cek invalid data
        invalid_ruangan = (mhs_data["ruangan"] == "INVALID").sum()
        null_mk = mhs_data["nama_mk"].isnull().sum()
        
        laporan_list.append({
            "Nama": nama,
            "Kelas": kelas,
            "Total SKS": int(total_sks),
            "Beban": beban,
            "Jumlah MK": len(mhs_data),
            "Alert": "Ada invalid data" if (invalid_ruangan > 0 or null_mk > 0) else "OK"
        })
    except Exception as e:
        print(f"Error processing mahasiswa {mhs_id}: {e}")

# Cetak laporan
laporan_df = pd.DataFrame(laporan_list)
print("\n" + "="*80)
print("LAPORAN AKADEMIK OTOMATIS")
print("="*80)
print(laporan_df.to_string(index=False))
print("\nSummary:")
print(f"Total Mahasiswa: {len(laporan_df)}")
print(f"Beban Ringan: {(laporan_df['Beban'] == 'Ringan').sum()}")
print(f"Beban Normal: {(laporan_df['Beban'] == 'Normal').sum()}")
print(f"Beban Berat: {(laporan_df['Beban'] == 'Berat').sum()}")
print(f"Ada Alert: {(laporan_df['Alert'] != 'OK').sum()}")
```
