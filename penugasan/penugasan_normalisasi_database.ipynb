# PENUGASAN: Normalisasi Database dari CSV

## Deskripsi Tugas

Anda akan bekerja dengan dataset Online Retail yang berisi transaksi penjualan online. Dataset ini adalah **single CSV file** dengan data yang belum ter-normalize.

### Tujuan Pembelajaran
- Load dan eksplorasi CSV file
- Validasi dan clean data (handle missing values, invalid data)
- **Normalize data** dari 1 CSV menjadi 3 tabel relational
- Create SQLite database dengan schema yang benar
- Insert normalized data ke database
- Write SQL queries untuk analisis

### Dataset Overview
**File:** `online_retail_dataset.csv`

**Kolom Original (Raw):**
- `InvoiceID` - ID transaksi unik
- `CustomerID` - ID pelanggan
- `ProductCode` - Kode produk
- `ProductName` - Nama produk
- `Quantity` - Jumlah item
- `UnitPrice` - Harga per unit
- `InvoiceDate` - Tanggal transaksi
- `Country` - Negara pelanggan

### Target Schema (Normalized)
Anda akan membuat 3 tabel:

**Tabel 1: customers**
```
customer_id (PK) | country
CUST2824         | Sweden
CUST1409         | Germany
```

**Tabel 2: products**
```
product_code (PK) | product_name     | unit_price
SKU416            | Wireless Mouse   | 147.35
SKU907            | USB-C Cable      | 395.07
```

**Tabel 3: orders**
```
invoice_id (PK) | customer_id (FK) | product_code (FK) | quantity | invoice_date
1               | CUST2824         | SKU416            | 5        | 2023-06-07
2               | CUST1409         | SKU907            | 3        | 2023-08-04
```
