### Query 3: Top 10 pelanggan berdasarkan total nilai pembelian [PALING BERAT]
Hitung total nilai pembelian per customer (quantity * unit_price). Tampilkan customer_id, negara, dan total nilai. Urut dari nilai terbesar.

**Konsep SQL yang digunakan:**
- JOIN antara 3 tabel: customers, orders, products
- GROUP BY untuk mengelompokkan per customer
- SUM + arithmetic (quantity * unit_price) untuk hitung total nilai
- ORDER BY + LIMIT untuk top 10

**Clue (Struktur Query):**
```
SELECT c.[kolom], SUM(o.[qty_kolom] * p.[price_kolom]) as total_value
FROM customers c
JOIN orders o ON c.customer_id = o.customer_id
JOIN products p ON o.product_code = p.product_code
GROUP BY c.customer_id
ORDER BY total_value DESC
LIMIT 10
```

```

WHERE c.customer_id IS NULL

Untuk cek duplikat:

FROM orders
GROUP BY invoice_id
HAVING cnt > 1
```
